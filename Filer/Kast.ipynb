# <font color=red>Ball i et loddrett kast</font>

```python
def y(g, v_0, t): #Definerer en funksjon y med variabler g, v_0 og t
  return v_0*t - 0.5*g*(t**2) #Returnerer verdien til formelen

g = 9.81 #Definerer en tilnærming av tyngdeakselerasjonen
v_0 = float(input('Skriv inn startfarten: ')) #Oppgaven sier 5
t = float(input('Skriv inn tiden: ')) #Oppgaven sier 0.6

Y = y(g, v_0, t) #Kaller på funksjonen y med bestemte variabler

print(f'Høyden etter {t} sekunder med startfarten {v_0} m/s er {Y:.2f} meter') #Printer ut en svar tekst med verdier
```

```python
import numpy as np #Importerer biblioteket numpy
import matplotlib.pyplot as plt #Importerer biblioteket matplotlib.pyplot
%config InlineBackend.figure_format = 'svg' #Setter formatet for aksene i de grafiske framstillingene

t = np.linspace(0, 1, 1000) #Fyller listen med 0 til 1 i 1000 steg

plt.plot(t, y(g, v_0, t)) #Kaller på definisjonen y med bestemte "konstanter" og én "variabel", og plotter den mot t
```

## <font color=blue>Ball i et skrått kast</font>

```python
import numpy as np
import matplotlib.pyplot as plt #Importerer biblioteket matplotlib.pyplot
%config InlineBackend.figure_format = 'svg' #Setter formatet for aksene i de grafiske framstillingene

def x(v_0, theta, t):
    return v_0*np.cos(theta*np.pi)*t

def y(g, v_0, theta, t):
    return v_0*np.sin(theta*np.pi)*t - 0.5*g*t**2

X = []
Y = []

g = 9.81 #Definerer en tilnærming av tyngdeakselerasjonen
v_0 = float(input('Skriv inn startfarten: ')) #Spør etter en verdi for startfarten
theta = float(input('Skriv inn vinkelen i radianer delt på pi: ')) #Spør etter utgangsvinkelen
t = np.linspace(0, 1, 1000) #Lager en liste over 1000 tidspunkt fra 0s til 1s

for k in range(len(t)):
    X += [x(v_0, theta, t[k])]
    Y += [y(g, v_0, theta, t[k])]
    
plt.plot(X,Y)
```
