```python
a = list(range(1,6,1))

#Her er a og b avhengig av hverandre
b = a
b[0] = 99
print(a, b)

#Her er a og c uavhengig av hverandre
c = a[:]
c[-1] = 99
print(a, c)
```

```python
from math import sqrt
import matplotlib.pyplot as plt #Importerer biblioteket matplotlib.pyplot
%config InlineBackend.figure_format = 'svg' #Setter formatet for aksene i de grafiske framstillingene

x = [0]*100
y = x[:]

for i in range(100):
    x[i] = i
    y[i] = sqrt(i)

plt.xlabel('$x$')      # Merker x-aksen
plt.ylabel('$y$')      # Merker y-aksen
plt.title('x mot y')   #Tittel
plt.plot(x, y, '.-', label='$y = sqrt(x)$')
plt.axhline(0, color='gray', lw=0.5) 
plt.legend()           # Infoboksen øverst i hjørnet. Bruker data fra label
plt.show()
```

## Fibonacci-følgen

```python
s = sum(1/i for i in range(1, 101)) #Løse en summasjon av følge
print("s = %g" % s)


#Fibonacci-følgen fra 1 til 10
F = [1]*10
for k in range(2, 10):
    F[k] = F[k-1] + F[k-2]
print(F)

#Enda en metode fo
F = [1, 1]
for k in range(2, 10):
    F.append(F[k-1] + F[k-2])
print(F)
```

```python
Cdegrees = list(range(-20, 41, 5))
Fdegrees = [(9/5)*C + 32 for C in Cdegrees]

table1 = [Cdegrees, Fdegrees] #To lister hvor table1[0][7] og table1[1][7] mat

table2 = [[C, F] for C, F in zip(Cdegrees, Fdegrees)] #Setter sammen matchende elementer fra to lister

print(table1)
print(table2)
```
